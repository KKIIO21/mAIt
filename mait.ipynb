```python
import os
import openai
import requests
import json
import playsound
import sounddevice as sd
from scipy.io.wavfile import write
from gtts import gTTS # gTTS install 필요


# voice record
def record_voice(title, duration=5, fs=44100):
    print("녹음을 시작하겠습니다.")
    myrecording = sd.rec(int(duration * fs), samplerate=fs, channels=2)
    sd.wait()  # 녹음이 끝날 때까지 기다립니다.
    write("./result" + title + '.wav', fs, myrecording)  # wav 파일로 저장합니다.
    
    
    
# voice to text
def voice2text(voiceFile, title):
    try:
        # API Endpoint
        url = "https://clovaspeech-gw.ncloud.com/recog/v1/stt"
        headers = {
            "X-CLOVASPEECH-API-KEY": "56f81be91e9344fd9825ff2dcb2b6d64", #콘솔 빌더의 비밀키
            "Content-Type": "application/octet-stream"
        }
        params = {
            "lang": "Kor"  # 언어 선택 (Kor, Eng, Jpn, Chn 중 선택)
        }

        # 음성 파일 입력
        # **확장자 확인**
        data = open(voiceFile, 'rb') # 음성 길이는 최대 60초(단문 인식)
        response = requests.post(url, headers=headers, params=params, data=data)
        rescode = response.status_code

        if(rescode == 200):
            result = response.json()
            text = result['text']
            print(f"입력: {text}")
            return text
        else:
            print("Error : " + response.text)
            return None
        
    except Exception as e:
        print(f"voice2text 실행 중 오류 발생: {str(e)}")
        return None

        
        
# text to voice
def text2voice(text, title):
    try:
        # TTS 객체 생성
        tts = gTTS(text=text, lang='ko')

        # mp3 파일로 저장
        tts.save("result/" + title + "_답변.mp3")
        print(f"출력: {text}")
        return True
        
    except Exception as e:
        print(f"text2voice 실행 중 오류 발생: {str(e)}")
        return False
    

def chatBot():
    try:
        title = 
        voiceFile = "./voice/" + title + ".wav"
        text = voice2text(voiceFile, title) # 음성 입력

        # OpenAI API 키 설정
        openai.api_key = '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'

        # 챗봇 세션 시작
        response = openai.ChatCompletion.create(
          model="gpt-3.5-turbo",  # 사용할 모델 설정
          messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": text},
            ]
        )

        r_text = response['choices'][0]['message']['content']
        text2voice(r_text, title)
        playsound.playsound("./result/"+ title + "_답변.mp3")
        return True
        
    except Exception as e:
        print(f"챗봇 실행 중 오류 발생: {str(e)}")
        return False

    
    
if __name__ == '__main__':
    if chatBot():
        print("chatBot completed.")
    else:
        print("chatBot failed.")
```
