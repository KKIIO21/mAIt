```python
# voice to text

import requests
import json

# API Endpoint
url = "https://clovaspeech-gw.ncloud.com/recog/v1/stt"

headers = {
    "X-CLOVASPEECH-API-KEY": "56f81be91e9344fd9825ff2dcb2b6d64", #콘솔 빌더의 비밀키
    "Content-Type": "application/octet-stream"
}

params = {
    "lang": "Kor"  # 언어 선택 (Kor, Eng, Jpn, Chn 중 선택)
}

# 음성 파일
# **확장자 확인**
data = open('voice/test3.mp3', 'rb') # 음성 길이는 최대 60초(단문 인식)

response = requests.post(url, headers=headers, params=params, data=data)
rescode = response.status_code

if(rescode == 200):
    result = response.json()
    text = result['text'] # text 부분만 str로 저장
    print("원본: 내가 그린 구름그림은 새털구름 그린 구름그림이고, 네가 그린 구름그림은 깃털구름 그린 구름그림이다.")
    print("결과:", text)
else:
    print("Error : " + response.text)
```

```python
# text to voice

import requests
import json

# API Endpoint
url = "https://clovaspeech-gw.ncloud.com/external/v1/voices"

headers = {
    "X-CLOVASPEECH-API-KEY": "56f81be91e9344fd9825ff2dcb2b6d64",  # 발급받은 Secret Key 입력
    "Content-Type": "application/json"
}

# 변환할 텍스트
data = {
    "voiceType": "WOMAN_READ_CALM",  # 발음자 선택, 여러가지 선택 가능
    "text": text,  # 음성으로 변환할 텍스트
    "language": "ko-KR",  # 언어 선택 (ko-KR, en-US 등)
    "audioEncoding": "MP3"  # 출력 포맷 선택 (MP3, OGG, PCM 등)
}

response = requests.post(url, headers=headers, data=json.dumps(data))
rescode = response.status_code

if(rescode == 200):
    print("TTS mp3 저장")
    with open('voice.mp3', 'wb') as f:
        f.write(response.content)
else:
    print("Error : " + response.text)
```
